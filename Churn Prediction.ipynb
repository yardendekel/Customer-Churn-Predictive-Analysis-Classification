
<head> <h1> <b> < Data Cleaning </b> </h1> </head>


```python
df = df.rename(columns = str.lower) # renames all the columns headers to lower letters
df
```

```python
df.columns
```

```python
df = df.replace(' ', np.nan) # replaces all empty values with NaN
nan_values = df[df.isna().any(axis=1)] # placement of the dataframe with all the rows that have NaN into a variable
nan_values
```

```python
df.isna().sum()
```

```python
df['totalcharges'] = df['totalcharges'].fillna(df['monthlycharges']) # replace all the NaN values in 'totalcharges' column with the 'monthlycharges' value respectively
df.loc[pd.Series(nan_values.index), ['monthlycharges','totalcharges']] # shows the difference at the rows which contained NaN
```

```python
df.dtypes
```

```python
df = df.drop(['customerid'], axis=1) # drops 'customerid' column
```

```python
df.loc[df['multiplelines'] == 'No phone service', 'multiplelines'] = 'No' # replace the 'No phone service' value with 'No' value at 'multiplelines' column
df.loc[df['internetservice'] == 'No', ['onlinesecurity', 'onlinebackup', 'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies']] = 'No' # replace the 'No internet service' value with 'No' value at all the columns mentioned
```

```python
df['totalcharges'] = df['totalcharges'].astype(float)
```

```python
df
```

```python
for column in df.columns: # turns all object datatype to an int datatype, all except columns which contain categorical values
    if (df[column].dtype == 'object') & (column != 'gender') & (column != 'internetservice') & (column != 'contract') & (column != 'paymentmethod'):
        df[column] = (df[column] == 'Yes').astype(int)
df
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df, columns=['gender', 'internetservice', 'contract', 'paymentmethod']) # split the categorical columns mentioned to new columns for each category
df.dtypes
```

```python
for column in df.columns: # turns all uint8 datatype to an int datatype
    if df[column].dtype == 'uint8':
        df[column] = df[column].astype(int)
df.dtypes
```

<head> <h1> <b>  Train Test Split </b> </h1> </head>

```python
# Create Train & Test Data
from sklearn.model_selection import train_test_split
X, y = df.drop('churn', axis=1), df['churn']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0, shuffle=True)
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

<head> <h1> <b>  Benchmark </b> </h1> </head>

```python
y_train.value_counts()
```

```python
def get_benchmark_predictions(x, benchmark_value): # The function returns a numpy array on the same length as x with all values equal to benchmark_value
    return np.ones(len(x))*benchmark_value

benchmark_value = 0

y_test_pred_Benchmark = get_benchmark_predictions(X_test, benchmark_value)
y_test_pred_Benchmark # Basicly, our benchmark predicts, always 0 (zero) --> no churn
```

<head> <h1> <b>  KNN Classification </b> </h1> </head>

```python
from sklearn.neighbors import KNeighborsClassifier  # Importing the algorithm
from sklearn.metrics import accuracy_score # importing "accuracy_score" from "sklearn.metrics"

train_acc = []
test_acc = []

for i in range(1,30):
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(X_train, y_train) # running\training the algorithm with the train data

    y_train_pred_Knn = clf.predict(X_train) # making a prediction based on "train" data features
    train_accuracy = accuracy_score(y_train, y_train_pred_Knn)
    train_acc.append(train_accuracy)

    y_test_pred_Knn = clf.predict(X_test) # making a prediction based on "test" data features
    test_accuracy = accuracy_score(y_test, y_test_pred_Knn)
    test_acc.append(test_accuracy)

fig = plt.figure(figsize=(16,9))
ax1 = fig.add_subplot(111)

ax1.scatter(range(1, 30), train_acc, s=10, c='b', marker="s", label='Training Accuracy')
ax1.scatter(range(1, 30) ,test_acc, s=10, c='r', marker="o", label='Testing Accuracy')
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Neighbors")
plt.legend(loc='upper left')
plt.show()
```

```python
clf = KNeighborsClassifier(n_neighbors=23)
clf.fit(X_train, y_train) # running\training the algorithm with the train data
y_test_pred_Knn = clf.predict(X_test) # making a prediction based on "test" data features

test_acc = accuracy_score(y_test, y_test_pred_Knn)
print(test_acc)

output = pd.DataFrame({'churn_what_actualy_happened':y_test, 'churn_predicted_by_model': y_test_pred_Knn}) # saving results to DataFrame
output
```

<head> <h1> <b>  Scaled KNN Classification </b> </h1> </head>


```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.transform(X_test)

train_acc = []
test_acc = []

for i in range(1,30):
    clf = KNeighborsClassifier(n_neighbors=i)
    clf.fit(x_train_scaled, y_train)

    y_train_pred_Knn_Scaled = clf.predict(x_train_scaled)
    train_accuracy = accuracy_score(y_train, y_train_pred_Knn_Scaled)
    train_acc.append(train_accuracy)

    y_test_pred_Knn_Scaled = clf.predict(x_test_scaled)
    test_accuracy = accuracy_score(y_test, y_test_pred_Knn_Scaled)
    test_acc.append(test_accuracy)

fig = plt.figure(figsize=(16,9))
ax1 = fig.add_subplot(111)

ax1.scatter(range(1, 30), train_acc, s=10, c='b', marker="s", label='Training Accuracy')
ax1.scatter(range(1, 30) ,test_acc, s=10, c='r', marker="o", label='Testing Accuracy')
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Neighbors")
plt.legend(loc='upper left')
plt.show()
```

```python
clf = KNeighborsClassifier(n_neighbors=10)
clf.fit(x_train_scaled, y_train)
y_test_pred_Knn_Scaled = clf.predict(x_test_scaled)

test_acc = accuracy_score(y_test, y_test_pred_Knn_Scaled)
print(test_acc)

output = pd.DataFrame({'churn_what_actualy_happened':y_test, 'churn_predicted_by_model': y_test_pred_Knn_Scaled}) # saving results to DataFrame
output
```

<head> <h1> <b>  Decision Tree Classification </b> </h1> </head>

```python
from sklearn.tree import DecisionTreeClassifier # Importing the algorithm

train_acc = []
test_acc = []

for i in range(1,30):
    clf = DecisionTreeClassifier(max_depth=i)
    clf.fit(X_train, y_train)

    y_train_pred_DecisionTree = clf.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred_DecisionTree)
    train_acc.append(train_accuracy)

    y_test_pred_DecisionTree = clf.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred_DecisionTree)
    test_acc.append(test_accuracy)

fig = plt.figure(figsize=(16,9))
ax1 = fig.add_subplot(111)

ax1.scatter(range(1, 30), train_acc, s=10, c='b', marker="s", label='Training Accuracy')
ax1.scatter(range(1, 30) ,test_acc, s=10, c='r', marker="o", label='Testing Accuracy')
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Max Depth")
plt.legend(loc='upper left')
plt.show()
```

```python
clf = DecisionTreeClassifier(max_depth=5)
clf.fit(X_train, y_train) # running\training the algorithm with the train data
y_test_pred_DecisionTree = clf.predict(X_test) # making a prediction based on "test" data features

test_acc = accuracy_score(y_test, y_test_pred_DecisionTree)
print(test_acc)

output = pd.DataFrame({'churn_what_actually_happened':y_test, 'churn_predicted_by_model': y_test_pred_DecisionTree}) # saving results to DataFrame
output
```

```python
# Plotting the decision tree
# Importing the necessary libraries
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from IPython.display import SVG
from graphviz import Source
from IPython.display import display

# this is afunction that we can always use for plotting decision trees, the function expects 3 arg as follows
def plot_tree(tree, features, labels):
    graph = Source(export_graphviz(tree, feature_names=features, class_names=labels, filled = True))
    display(SVG(graph.pipe(format='svg')))

# Using the function above, with the 3 arg - tree, feaures, labels
plot_tree(clf, X_train.columns, ['No Churn', 'Churn'])
```

<head> <h1> <b> Random Forest Classification </b> </h1> </head>

```python
from sklearn.ensemble import RandomForestClassifier # Importing the algorithm

train_acc = []
test_acc = []

for i in range(1,30):
    clf = RandomForestClassifier(n_estimators=100, max_depth=i, random_state=0)
    clf.fit(X_train, y_train)

    y_train_pred_RandomForest = clf.predict(X_train)
    train_accuracy = accuracy_score(y_train, y_train_pred_RandomForest)
    train_acc.append(train_accuracy)

    y_test_pred_RandomForest = clf.predict(X_test)
    test_accuracy = accuracy_score(y_test, y_test_pred_RandomForest)
    test_acc.append(test_accuracy)

fig = plt.figure(figsize=(16,9))
ax1 = fig.add_subplot(111)

ax1.scatter(range(1, 30), train_acc, s=10, c='b', marker="s", label='Training Accuracy')
ax1.scatter(range(1, 30) ,test_acc, s=10, c='r', marker="o", label='Testing Accuracy')
ax1.set_ylabel("Accuracy")
ax1.set_xlabel("Max Depth")
plt.legend(loc='upper left')
plt.show()
```

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=9, random_state=0)
clf.fit(X_train, y_train)
y_test_pred_RandomForest = clf.predict(X_test)

test_acc = accuracy_score(y_test, y_test_pred_RandomForest)
print(test_acc)

output = pd.DataFrame({'churn_what_actually_happened':y_test, 'churn_predicted_by_model': y_test_pred_RandomForest})
output
```
